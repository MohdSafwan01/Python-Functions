generally we do this

```python
def myfunc(a,b):
  return sum((a,b))*0.05
```

```python
myfunc(40,60)
```

but the issue here is , if we need to add more parameters inside the function we have to do it manually like this

```python
def myfunc(a,b,c=0,d=0,e=0):
  return sum ((a,b,c,d,e))*0.05
```

```python
myfunc(40,60)
```

```python
myfunc(40,60,100)
```

```python
myfunc(40,60,100,100,80)
```

but when we try to insert more inputs which exceed the no. of parameters in the function definition it will give error like this:-

```python
myfunc(40,60,100,100,100,100)
```

"TypeError: myfunc() takes from 2 to 5 positional arguments but 6 were given"

this is the error we will get

but for this issue we can use '*args'(arguements)

this allows us to enter multiple no. of inputs without needing to define in the function definition

*args acts as a tuple

```python
def myfunc(*args):
  return sum(args)*0.05
```

```python
myfunc(40,30)
```

```python
myfunc(40,60,100,100,100,100)
```

```python
def myfunc(*args):
  print(args)
```

```python
myfunc(40,40,50,60,100,100)
```

```python
def myfunc(*args):
  for item in args:
    print(item)
```

```python
myfunc(40,40,50,60,100,100)
```

2nd is **kwargs(key word arguements)

**kwargs differ from *args because they acts as a dictionary in the function and *args acts a tuple



```python
def myfucn(**kwargs):
  if 'fruit' in kwargs:
    print('my fruit of choice is{}'.format(kwargs['fruit']))
  else:
    print('I did not found any fruit')
```

```python
myfucn(fruit =' apple', veggie = 'tomato')
```

using *args and **kwargs

```python
def myfunc(*args,**kwargs):
  print('I would like {} {}'.format(args[0], kwargs['food']))
```

```python
myfunc(10,20,30,40,food='apple')
```

```python
def myfunc(*args, **kwargs):
    print(args)
    print(kwargs)
    print('I would like {} {}'.format(args[0], kwargs['food']))
```

```python
myfunc(10, 20, 30, fruit='orange', food='eggs', animal='dog')
```
