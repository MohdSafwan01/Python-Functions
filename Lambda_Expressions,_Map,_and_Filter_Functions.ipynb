```python
def square(num):
  return num**2
```

```python
square(3)
```

```python
my_nums=[1,2,3,4,5]
```

```python
for item in map(square,my_nums):
  print(item)
```

```python
list(map(square,my_nums))
```

```python
def splicer(mystring):
  if len(mystring)%2==0:
    return "EVEN"
  else:
    return mystring[0]
```

```python
mystring = ['Andy','Eve','Sally','Adam']
```

```python
list(map(splicer,mystring))
```

Filter function

```python
def check_even(mynums):
  return mynums%2 == 0
```

```python
mynums = [1,2,3,4,5,6]
```

```python
list(filter(check_even,mynums))
```

```python
for n in filter(check_even,mynums):
  print(n)
```
