Question: Password Validator Game
Create a password validation system where:

The system has a secret password stored

The user gets 3 attempts to guess it

After each wrong guess, it tells them how many attempts remain

If they guess correctly, they "unlock" the system

If they fail all 3 attempts, the system "locks"

```python
def set_password():
  return ('python123')
```

```python
  def get_user_input():

   passsword = input('enter password')
   return passsword

```

```python
def validate_password(secret_pass, user_pass, attempts_left):

  if secret_pass == user_pass:
    print('Correct Password')
    return True
  else:
    print('wrong password')
    print(str(attempts_left)+ " attempts remaining")
    return False
```

```python

def run_game():
  secret_pass = set_password()
  attempts_left = 3
  for i in range(3):


    user_input = get_user_input()
    result = validate_password(secret_pass, user_input, attempts_left)

    if result == True:
      break
    else:
      attempts_left=attempts_left-1
  print('system locked')





```

```python
run_game()
```
