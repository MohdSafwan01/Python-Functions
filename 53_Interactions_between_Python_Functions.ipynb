a code example of the famous game "three cup monte"

```python
my_list = [' ','O',' ']
```

creating a shuffle function

```python
def shuffle_list(my_list):
  shuffle(my_list)
  return my_list
```

taking users guess as input

```python
def user_guess():
  guess = ''
  while guess not in ['0','1','2']:
   guess = input("Pick a number 0,1 or 2")
  return int(guess)
```

checking the user guess

```python
def check_guess(my_list,guess):
  if my_list[guess] == 'O':
    print("Correct")
  else:
    print("Wrong Guess")
    print(my_list)
```

combining all the functions

```python
my_list = [' ','O',' ']

mixed_list = shuffle_list(my_list)

guess = user_guess()

check_guess(mixed_list,guess)
```
