Write a function that checks whether a number is in a given range (inclusive of high and low)

```python
def ran_check(num,low,high):
  if low<num<high+1:
    print('{} is in the range between {} and {}'.format(num,low,high))
  else:
    print('{} is not in range between {} and {}'.format(num,low,high))
```

```python
ran_check(5,2,7)
```

```python
ran_check(1,2,5)
```

```python
def ran_bool(num,low,high):
  if low<num<high+1:
    return True
  else:
    return False
```

```python
ran_bool(3,1,10)
```

```python
ran_bool(1,3,10)
```

```python
def up_low(s):
  count_upper = 0
  count_lower = 0
  for i in s:
    if i.isupper():
      count_upper=count_upper+1

    if i.islower():
      count_lower=count_lower+1


  print("No. of Upper case characters :{}".format(count_upper))
  print("No. of Lower case Characters :{}".format(count_lower))
```

```python
up_low('Hello Mr. Rogers, how are you this fine Tuesday?')
```

```python
up_low('Hi my name is Safwan snd im on my journey tp Learn Python from Scratch')
```

```python
def unique_list(lst):
  unq_list = []
  for i in lst:
    if i not in unq_list:
      unq_list.append(i)
  return unq_list
```

```python
unique_list([1,1,1,1,2,2,3,3,3,3,4,5])
```

```python
def palindrome(s):
  reversed_s = s[::-1]
  if s == reversed_s:
    return True
  else :
    return False
```

```python
palindrome('helleh')
```

```python
palindrome('madam')
```

```python
palindrome('karan')
```
